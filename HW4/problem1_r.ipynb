1. First I import and read the adult_training.csv file. I then use it to create the dataframe trainingDF and display it. In order to import the data I had to upload the adult_training.csv file to the session storage before I could read it and create a dataframe

```r
#Import data and create dataframe
data <- read.csv("adult_training.csv")
trainingDF <- data
head(trainingDF, 100)
```

2. After creating and plotting the contingency table for work class and sex, I analyzed the results to see if I could draw any conclusions. I found that the majority of the people in this dataset (69.4%) belong to the private work class. Another thing I noticed was that their are more males than females in the dataset. The data is 66.8% male. This can explain why there are more males than females for every different work class.

```r
install.packages("gmodels")
#Creates contingency table for work class and sex
library(gmodels)
CrossTable(trainingDF$workclass, trainingDF$sex)
```

```r
library("ggplot2")
#Creates a plot of the contigency table for work class and sex
P <- prop.table(table(trainingDF$workclass, trainingDF$sex), margin=2)
ggplot(as.data.frame(P), aes(x = 'workclass', y = Freq, fill = Var1)) +
geom_bar(stat="identity", position = "dodge")
```

3. After creating and plotting the contingency table for marital status and sex, I analyzed the results to see if I could draw any conclusions. While theree are still more males than females like the last contingency table there are still conclusion that can be drawn based on sex. While the most common marital status for males was being married to a civilian spouse (60.7% of males), the most common marital status for females was never being married (44% of females).

4. To create a dataframe with adults over the age of 40 I created a new dataframe that is equal to the subset of the original dataframe where age is greater than 40. I then fix the indexes and display the values of the new dataframe.

```r
#Creates new df adultOver40DF by taking the subset of rows with ages above 40 in trainingDF
adultOver40DF <- subset(trainingDF, age > 40)
#Fixes the indexes for the new df and prints it
rownames(adultOver40DF) <- NULL
head(adultOver40DF, 100)
```

6. I looked for outliers for education by creating a new colmun scaled education and populating it with each education's z score. I then printed the amount of rows that exist in the subsets where scaled education is greater than 3 and scaled education is less than -3. I found that there were 113 outliers for education.

```r
#Creates new coloumn scaled education and fill it with each education's respective z score
trainingDF$scaled_education <- scale(x = trainingDF$education)
#Checks for outliers by displaying the amount of rows in the subsets where scaled education is greater than 3 or less than -3
nrow(subset(trainingDF, scaled_education > 3)) + nrow(subset(trainingDF, scaled_education < -3))
```

7. First I standardized the age variable by creating a new column scaled age and populating it with the z scores of the different ages. I then checked for outliers by printing the amount of rows that exist in the subsets where scaled age is greater than 3 and scaled age is less than -3. I found that there were 60 outliers and that the most extreme outlier was 90 years old.


```r
#Creates new coloumn scaled age and fill it with each age's respective z score
trainingDF$scaled_age <- scale(x = trainingDF$age)
#Checks for outliers by displaying the amount of rows in the subsets where scaled age is greater than 3 or less than -3 (60)
nrow(subset(trainingDF, scaled_age > 3)) + nrow(subset(trainingDF, scaled_age < -3))
#Displays the most extreme outlier (90)
max(subset(trainingDF, scaled_age > 3)$age)
```
